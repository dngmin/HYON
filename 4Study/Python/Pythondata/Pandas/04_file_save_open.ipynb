# 4. 파일 저장 및 열기
DataFrame 객체를 excel,csv,txt등의 형태의 파일로 저장 및 열기

```python
import pandas as pd
data={
    '日本語' : ['リンゴ','オレンジ','パイナップル','イチゴ','ナシ'],
    'English' : ['apple','orange','pineapple','strawberry','pear'],
    '한글' : ['사과','오랜지','파인애플','딸기','배'],
    'kcal' : ['54','47.1','50.9','33.9','57.1']
}

df = pd.DataFrame(data,index=['no1','no2','no3','no4','no5'])
df.index.name='no'
df
```

## 저장하기
### csv 파일로 저장

```python
df.to_csv('fruit.csv',encoding='utf-8-sig')
```

```python
df.to_csv('fruit.csv',encoding='utf-8-sig',index=False)
```

### 텍스트(.txt)파일로 저장

```python
df.to_csv('fruit.txt',sep='\t') #tab으로 구분된 텍스트 파일
```

### 엘셀 파일로 저장

```python
df.to_excel('fruit.xlsx')
```

***
## 열기
### csv 파일열기

```python
df=pd.read_csv('fruit.csv')
df
```

```python
df=pd.read_csv('fruit.csv',skiprows=1)  #지정된 갯수 만큼의 row를 건너뜀
df
```

```python
df=pd.read_csv('fruit.csv',skiprows=[1,3,5]) #지정된 row 제외
df
```

```python
df=pd.read_csv('fruit.csv',nrows=2) #지정된 갯수(제목비포함)만큼의 row만
df
```

```python
df=pd.read_csv('fruit.csv',skiprows=2,nrows=2) #첫 2 row 무시, 이후 2row
df
```

### 텍스트(.txt) 파일 열기

```python
df=pd.read_csv('fruit.txt',sep='\t')
df
```

```python
df=pd.read_csv('fruit.txt',sep='\t',index_col='no')
df
```

```python
df=pd.read_csv('fruit.txt',sep='\t')
df.set_index('no',inplace=True)
df
```

### 엘셀 파일 열기

```python
df=pd.read_excel('fruit.xlsx')
df
```

```python
df=pd.read_excel('fruit.xlsx',index_col='no')
df
```
