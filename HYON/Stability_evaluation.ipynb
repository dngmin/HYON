# ランチクリア安定性評価
---

* alph = tan-1((w * sinθ)/(v - w * cosθ))
* beta = sin-1(w / sqrt(w^2 + v^2))

    * alph = 順風迎角
    * beta = 横滑り角
    * θ = 打上射角
    * v = ランチクリア速度
    * w = 評定地上速度

```python
import numpy as np

θ = 80 * np.pi / 180
v = 28.81
w = 10

a = np.arctan((w * np.sin(θ)) / (v - w * np.cos(θ))) * 180 / np.pi
b = np.arcsin(w / (np.sqrt(w**2 + v**2))) * 180 / np.pi

print(f'a = {a}')
print(f'b = {b}')
```

# 飛行中の安定性評価
---

* Fst = (Xcg - Xcp) / L * 100%
* Cs = (Xcg - Xcp) / D
    * Xcg = 機体後端からの重心位置
    * Xcp = 機体後端から圧力中心位置
    * L = 全長(ピトー管およびフィンを含む)
    * D = 代表直径

```python
Xcg_pre = 611
Xcg_se = 620
Xcp = 406
L = 1350
D = 102

Fst_pre = (Xcg_pre - Xcp) / L *100
Fst_se = (Xcg_se - Xcp) / L *100
Cs_pre = (Xcg_pre - Xcp) / D
Cs_se = (Xcg_se - Xcp) / D

print(f'Fst_pre = {Fst_pre}')
print(f'Fst_se = {Fst_se}')
print(f'Cs_pre = {Cs_pre}')
print(f'Cs_se = {Cs_se}')
```
