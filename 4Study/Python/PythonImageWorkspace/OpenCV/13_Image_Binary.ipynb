# 13. 이미지 변형 (이진화)

## Threshold (임계값)

```python
import cv2
img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img,thresh=127,maxval=255,type=cv2.THRESH_BINARY)

cv2.imshow('img',img)
cv2.imshow('binary',binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Trackbar (값 변화에 따른 변형 확인)

```python
import cv2

def empty(pos):
    # print(pos)
    pass

img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE)

cv2.namedWindow('Trackbar')

cv2.createTrackbar('threshold','Trackbar',127,255,empty) # bar이름, 창이름, 초기값, 최대값, 이벤트처리

while True:
    thresh = cv2.getTrackbarPos('threshold','Trackbar') # bar이름, 창 이름
    ret, binary = cv2.threshold(img,thresh,255,cv2.THRESH_BINARY)

    if not ret:
        break

    cv2.imshow('Trackbar',binary)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import cv2
img = cv2.imread('threshold.png',cv2.IMREAD_GRAYSCALE)

ret, binary1 = cv2.threshold(img,thresh=0,maxval=255,type=cv2.THRESH_BINARY) # thresh 0 초과 - 흰색처리
ret, binary2 = cv2.threshold(img,thresh=127,maxval=255,type=cv2.THRESH_BINARY) # thresh 127 초과 - 흰색처리
ret, binary3 = cv2.threshold(img,thresh=195,maxval=255,type=cv2.THRESH_BINARY) # thresh 195 초과 - 흰색처리

cv2.imshow('img',img)
cv2.imshow('binary1',binary1)
cv2.imshow('binary2',binary2)
cv2.imshow('binary3',binary3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Adaptive Threshold
이미지를 작은 영역으로 나누어서 임계치 적용

```python
import cv2

def empty(pos):
    # print(pos)
    pass

img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE)

cv2.namedWindow('Trackbar')

# bar이름, 창이름, 초기값, 최대값, 이벤트처리
cv2.createTrackbar('block_size','Trackbar',25,100,empty) # 홀수만 가능, 1보다는 큰값Q   
cv2.createTrackbar('c','Trackbar',3,10,empty) #일반적으로 양수의 값을 사용

while True:
    block_size = cv2.getTrackbarPos('block_size','Trackbar') # bar이름, 창 이름
    c = cv2.getTrackbarPos('c','Trackbar')

    if block_size <=1: # 1이하면 3으로
        block_size = 3
    
    if block_size % 2 == 0: # 짝수면 홀수로 
        block_size += 1

    binary = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,block_size,c)

    cv2.imshow('Trackbar',binary)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

## 오츠 알고리즘

Bimadal image에 사용하기 적합 (최적의 임계치를 자동으로 발견)

```python
import cv2
img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret, otsu = cv2.threshold(img,-1,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print('otsu threshold',ret)

cv2.imshow('img',img)
cv2.imshow('binary',binary)
cv2.imshow('otsu',otsu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
