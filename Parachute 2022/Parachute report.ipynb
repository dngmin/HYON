### 1. 実験の予測値

```python
# library
import numpy as np
import pandas as pd
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
i = None
```

```python
# parameter and fomula for graph
rho = 1.166 # 20℃における空気密度 [kg/m^3]
A = 0.4693 # パラシュート投影面積 [m^2]
Cd = 0.70 # パラシュート抗力係数 [-]
g = 9.81 # 重力加速度
m = 1.306 # 機体質量 [kg]
dt = 0.04 # 時間刻み [s]
h = 15.0 # 落下支点高さ [m]
Vwind = 0.0 # 風(Vwind,0,0)

data = {
    't' : [],
    'x' : [],
    'y' : [],
    'vx' : [],
    'vy' : [],
    'ax' : [],
    'ay' : []
}

x= 0
y = h
vx = 0
vy = 0
ax = 0
ay = -g

for t in np.arange(0,100,dt):

    data['t'].append(t)
    data['x'].append(x)
    data['y'].append(y)
    data['vx'].append(vx)
    data['vy'].append(vy)
    data['ax'].append(ax)
    data['ay'].append(ay)

    x = x + vx * dt
    y = y + vy * dt
    vx = vx + ax * dt
    vy = vy + ay * dt
    ax = 0.5 * rho * ((vx-Vwind)**2) * A *Cd / m
    ay = -g + 0.5 * rho * (vy**2) * A * Cd / m

    if i == 0:
        break

    if y < 0:
        i = 0

df = pd.DataFrame(data)
df
```

```python
# graph
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import patches

X = np.array(data['x'])
Y = np.zeros(len(data['x']))
Z = np.array(data['y'])

fig = plt.figure()
ax = Axes3D(fig)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.plot(X,Y,Z)

plt.show()
```
