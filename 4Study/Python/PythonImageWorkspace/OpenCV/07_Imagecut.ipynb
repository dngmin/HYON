# 7. 이미지 자르기

영역을 잘라서 새로운 윈도우(창)에 표시

```python
import cv2
img = cv2.imread('img.jpg')
# img.shape # (640,526,3)

crop = img[250:350,100:300] # [세로범위 250~350 , 가로범위 10~300]
cv2.imshow('img',img) # 원본 이미지
cv2.imshow('crop',crop) # 잘린 이미지
cv2.waitKey(0)
cv2.destroyAllWindows()
```

영역을 잘라서 기존 윈도우에 표시

```python
import cv2
img = cv2.imread('img.jpg')

crop = img[250:350,100:300] # [세로범위 250~350 , 가로범위 100~100]
img[250:350,250:450] = crop

cv2.imshow('img',img) # 원본 이미지
cv2.waitKey(0)
cv2.destroyAllWindows()
```

in video

```python
import cv2

cap = cv2.VideoCapture('video.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    crop = frame[40:200,240:400]
    frame[40:200,80:240] = crop

    cv2.imshow('video',frame)
    # cv2.imshow('crop',crop)
    if cv2.waitKey(1) == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
