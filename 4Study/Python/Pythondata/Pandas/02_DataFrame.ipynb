2차원 데이터(Series들의 모음)

## Data 준비
사전 (dict) 자료구조를 통해 생성

예) 인물 데이터

```python
data = {
    '이름' : ['김동민','kimdongmin','キムドンミン'],
    '언어' : ['한글','English','日本語'],
    '글자수' : [2,7,3]
}

data
```

```python
data['이름']
```

```python
data['글자수']
```

## DataFrame 객체생성

```python
import pandas as pd
df=pd.DataFrame(data)
```

```python
df
```

## 데이터 접근

```python
df['이름']
```

```python
df['글자수']
```

```python
df[['이름','언어']]
```

## DataFrame 객체 생성 (index지정)

```python
df=pd.DataFrame(data,index=['a','b','c'])
df
```

## DataFrame 객체생성 (Column지정)
data 중에서 원하는 column만 선택하거나, 순서 병경 가능

```python
df=pd.DataFrame(data,columns=['이름','글자수'])
df
```

```python
df=pd.DataFrame(data,columns=['글자수','이름'])
df
```
