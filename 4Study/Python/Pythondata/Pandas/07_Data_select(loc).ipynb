# 7. 데이터 선택 (Location)
이름을 이용해서 원하는 row에서 원하는 col 선택

```python
import pandas as pd
df= pd.read_excel('fruit.xlsx',index_col='no')
df
```

```python
df.loc['no1'] # index 1번에 해당하는 전체 데이터
```

```python
df.loc['no5'] # index 5번
```

```python
df.loc['no1','English'] # index 1에 해당하는 English 데이터
```

```python
df.loc[['no1','no4'],'日本語'] # index 1,4번에 해당하는 日本語 데이터
```

```python
df.loc[['no2','no3'],['한글','kcal']] # index 2,3번에 해당하는 한글 kcal 데이터
```

```python
df.loc['no1':'no4','日本語':'English'] # index 1~4까지,日本語부터 English까지
```
