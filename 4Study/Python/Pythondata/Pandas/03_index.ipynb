# 3.Index
데이터에 접근할 수 있는 주소 값

```python
import pandas as pd
data={
    '日本語' : ['リンゴ','オレンジ','パイナップル','イチゴ','ナシ'],
    'English' : ['apple','orange','pineapple','strawberry','pear'],
    '한글' : ['사과','오랜지','파인애플','딸기','배']
}

df = pd.DataFrame(data,index=['no1','no2','no3','no4','no5'])
df
```

```python
df.index
```

## Index 이름설정

```python
df.index.name='서열'
df
```

## Index 초기화

```python
df.reset_index()
```

```python
df.reset_index(drop=True) # 원래쓰던 index 삭제
```

```python
df
```

```python
df.reset_index(drop=True,inplace=True)  #실제 데이터에 바로반영
df
```

## Index 설정
지정한 column으로 index 를 설정

```python
df.set_index('日本語')
```

```python
df
```

```python
df.set_index("한글",inplace=True)
df
```

## Index 정렬
Index 를 기준으로 오름차순, 내림차순 정렬

```python
df.sort_index() #Index 오름차순 정렬
```

```python
df.sort_index(ascending=False) #Index 내림차순 정렬
```
