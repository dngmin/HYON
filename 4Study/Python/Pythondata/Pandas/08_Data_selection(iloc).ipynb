# 8. 데이터 선택 (integer location)
위치를 이용하여 원하는 row 에서 원하는 col 선택

```python
import pandas as pd
df= pd.read_excel('fruit.xlsx',index_col='no')
df
```

```python
df.iloc[0] # 0번쨰 위치의 데이터
```

```python
df.iloc[4] # 4번째 위치의 데이터
```

```python
df.iloc[0:3] # 0~2번쨰 위치의 데이터
```

```python
df.iloc[0,1] # 0번쨰 위치의 1번째(English) 데이터
```

```python
df.iloc[4,3] # no5의 kcal 데이터( 4번쨰 위치의 3번쨰(kcal) 데이터)
```

```python
df.iloc[[0,1],3] #0,1 번쨰 위치의 3번째(kcal) 데이터
```

```python
df.iloc[[0,1],[1,2]] # 0,1번쨰 위치의 1,3번쨰(日本語,한글) 데이터
```

```python
df.iloc[0:4,2:3] # 0~3번쨰 위치의 2~3번째(한글.kcal) 데이터
```
